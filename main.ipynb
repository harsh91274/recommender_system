```python
movies=movies[['id', 'title', 'overview', 'genre']]
print(movies.head())
```

```python
movies['tags']=movies['overview']+movies['genre']
```

```python
movies
```

```python
new_data=movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

use TFIDF for converting the tags into vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english', max_features=10000)
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
#get list of similarity values for a particular movie
#enumerate gives you the index of the output as well 
#key provides to the column to sort on - here it is the value of the similairyt and not the index

sorted(list(enumerate(similarity[1])), reverse=True, key=lambda vector: vector[1])

```

```python
distance=sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0] #get index of input movie
    distance=sorted(list(enumerate(similarity[index])), reverse=True, key= lambda vector:vector[1])

    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
